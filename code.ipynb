```python
import pandas as pd

# Load Excel file (change the path if needed)
df = pd.read_excel("/content/Dataset.xlsx", sheet_name='Form Responses 1')

# Clean column names
df.columns = (
    df.columns.str.strip()
    .str.lower()
    .str.replace(r'\[.*?\]', '', regex=True)
    .str.replace(r'[^\w\s]', '', regex=True)
    .str.replace(r'\s+', '_', regex=True)
)

# Clean values
df['gender'] = df['gender'].str.strip().str.title()
df['age'] = df['age'].astype(str).str.strip()

# Add easier column names
df['mental_health_diagnosis'] = df['have_you_ever_been_diagnosed_with_a_mental_health_condition_by_a_professional_doctor_therapist_etc']
df['treatment'] = df['have_you_ever_received_treatmentsupport_for_a_mental_health_problem']

# Drop timestamp if exists
if 'timestamp' in df.columns:
    df.drop(columns=['timestamp'], inplace=True)

# Rename stress columns
stress_cols = [col for col in df.columns if 'what_is_your_stress_level' in col]
df.rename(columns={col: f"stress_{i+1}" for i, col in enumerate(stress_cols)}, inplace=True)
```

```python
# 1. Gender distribution
plt.figure()
sns.countplot(data=df, x="gender", order=df["gender"].value_counts().index)
plt.title("Distribution of Respondents by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Mental health diagnosis vs treatment
plt.figure()
sns.countplot(data=df, x="mental_health_diagnosis", hue="treatment")
plt.title("Mental Health Diagnosis vs. Treatment")
plt.xlabel("Diagnosed with Mental Health Condition")
plt.ylabel("Number of Respondents")
plt.legend(title="Received Treatment")
plt.tight_layout()
plt.show()

# 3. Average stress level per factor
stress_cols = [col for col in df.columns if 'stress_' in col]
df_stress_numeric = df[stress_cols].apply(pd.to_numeric, errors='coerce')

mean_stress = df_stress_numeric.mean().reset_index()
mean_stress.columns = ['Stress Factor', 'Average Level']

plt.figure()
sns.barplot(data=mean_stress, x='Stress Factor', y='Average Level', palette="coolwarm")
plt.title("Average Stress Levels by Factor")
plt.xlabel("Stress Factor")
plt.ylabel("Average Stress Level (1=Low, 5=High)")
plt.tight_layout()
plt.show()
```
